# First ML Model
Train a simple Random Forest model on the system metrics dataset.

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

df = pd.read_csv('../datasets/system_metrics_sample.csv')
X = df[['cpu', 'memory', 'latency']]
y = df['failure']
model = RandomForestClassifier(random_state=42)
model.fit(X, y)
```

```python
model.predict([[90, 84, 105]])
```
