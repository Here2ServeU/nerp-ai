# Anomaly Detection
Use Isolation Forest to detect unusual telemetry behavior.

```python
from sklearn.ensemble import IsolationForest
import pandas as pd

df = pd.read_csv('../datasets/system_metrics_sample.csv')
X = df[['cpu', 'memory', 'latency']]
detector = IsolationForest(contamination=0.2, random_state=42)
detector.fit(X)
```

```python
detector.predict([[99, 97, 145]])
```
